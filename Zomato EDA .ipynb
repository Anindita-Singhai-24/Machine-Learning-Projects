```python
temp=zomato.highlights[0]
temp
```

```python
def highlights(x):
    x=x.split("'")
    y=[]
    for i in range(0,len(x)):
        if i%2!=0:
            y.append(x[i])
    return y
highlights(temp)
```

```python
zomato.highlights=zomato.highlights.apply(highlights)
```

```python
zomato['highlights_count']=zomato.highlights.apply(lambda x:len(x))
```

```python
total_highlights= set()
for i in zomato.highlights:
    total_highlights.update(i)
```

#### Given the above analysis, it is clear that for different resturants which do not have a establishment type, we can replace them with the mode in price range 1 and 2 and drop them in price range 3 and 4


```python
p1=zomato[((zomato.establishment=='[]') & (zomato.price_range==1))].establishment.index
p2=zomato[((zomato.establishment=='[]') & (zomato.price_range==2))].establishment.index
```

```python
for i in p1:
    zomato['establishment'][i]="['Quick Bites']"
for i in p2:
    zomato['establishment'][i]="['Casual Dining']"
```

```python
zomato=zomato.drop(zomato[zomato.establishment=='[]'].index)
```

```python
def cuisines(x):
    x=x.split(",")
    return x
cuisines(zomato.cuisines[0])
```
